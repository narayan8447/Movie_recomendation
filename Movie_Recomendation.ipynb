```python
# Make sure you have this import at the top of your script
from thefuzz import process

def recommend(movie):
    # 1. --- FUZZY MATCHING LOGIC ---
    # Get a list of all movie titles from your DataFrame
    all_titles = new_df['title'].tolist()
    
    # Find the best match for the user's input from the list of all titles
    # It returns a tuple: (best_match_title, similarity_score)
    best_match = process.extractOne(movie, all_titles)
    
    # Check if the similarity score is high enough (e.g., above 80%)
    if best_match[1] < 80:
        print(f"Sorry, couldn't find a close match for '{movie}'. Please try again.")
        return

    # If the match is good, use the title it found
    matched_title = best_match[0]

    # --- The rest of the logic remains mostly the same ---

    # 2. Get the index of the movie that was matched
    movie_index = new_df[new_df['title'] == matched_title].index[0]

    # Get the similarity scores for that movie
    distances = similarity[movie_index]
    # Sort the movies by similarity and get the top 5
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    # 3. Print the recommendations, informing the user of the match
    print(f"Showing results for '{matched_title}' (based on your search for '{movie}'):")
    print("-" * 20) # Adding a separator line for clarity
    for i in movies_list:
        # i[0] is the index of the recommended movie
        print(new_df.iloc[i[0]].title)
```

```python
# Make sure you have this import at the top of your script
from thefuzz import process

def get_recommendations_for_evaluation(movie, n=5):
    """
    This is your recommendation logic, modified to RETURN a list for evaluation.
    """
    all_titles = new_df['title'].tolist()
    best_match = process.extractOne(movie, all_titles)
    
    # If no good match is found, return an empty list
    if best_match[1] < 80:
        return []

    matched_title = best_match[0]
    movie_index = new_df[new_df['title'] == matched_title].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:n+1]

    # --- KEY CHANGE IS HERE ---
    # Instead of printing, we create a list and return it
    recommendation_list = []
    for i in movies_list:
        recommendation_list.append(new_df.iloc[i[0]].title)
    
    return recommendation_list
```

```python
# Helper function to get the genres for a given movie
def get_genres(movie_title, df):
    # This assumes your original DataFrame is named 'Movies'
    return df[df['title'] == movie_title]['genres'].values[0]

def evaluate_recommender(test_movie, original_df, n=10):
    """
    Calculates Precision and Recall for a single movie's recommendations.
    """
    # 1. Get the list of recommended movies using our new function
    recommended_movies = get_recommendations_for_evaluation(test_movie, n=n)
    if not recommended_movies:
        return 0, 0

    # 2. Define the Ground Truth (relevant items based on genre)
    test_movie_genres = set(get_genres(test_movie, original_df))
    relevant_movies = set()
    for index, row in original_df.iterrows():
        if row['title'] == test_movie:
            continue
        movie_genres = set(row['genres'])
        if test_movie_genres.intersection(movie_genres):
            relevant_movies.add(row['title'])

    # 3. Calculate Metrics
    true_positives = set(recommended_movies).intersection(relevant_movies)
    precision = len(true_positives) / len(recommended_movies)
    
    if not relevant_movies:
        return precision, 0
    recall = len(true_positives) / len(relevant_movies)
    
    return precision, recall
```

```python
# Run evaluation on a sample of 100 movies
sample_size = 100
movie_sample = new_df['title'].sample(n=sample_size, random_state=42)

total_precision = 0
total_recall = 0

# IMPORTANT: Make sure 'Movies' is your original DataFrame with the 'genres' column
original_movies_df = Movies 

for movie in movie_sample:
    precision, recall = evaluate_recommender(movie, original_movies_df, n=10)
    total_precision += precision
    total_recall += recall

# Calculate the average scores
avg_precision = total_precision / sample_size
avg_recall = total_recall / sample_size

print(f"--- Evaluation Results (based on a sample of {sample_size} movies) ---")
print(f"Average Precision@10: {avg_precision:.4f}")
print(f"Average Recall@10:    {avg_recall:.4f}")
```
